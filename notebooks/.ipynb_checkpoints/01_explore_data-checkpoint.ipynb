```python
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
base_dir = "../data/chest_xray"

for split in ["train","test","val"]:
    for label in ["NORMAL","PNEUMONIA"]:
        path = os.path.join(base_dir, split, label)
        count = len(os.listdir(path))
        print(f"{split}/{label}: {count} images")
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15, 5))
normal_path = os.path.join(base_dir,"train","NORMAL")
pneumonia_path = os.path.join(base_dir,"train","PNEUMONIA")

normal_img = os.listdir(normal_path)[0]
pneumonia_img = os.listdir(pneumonia_path)[0]

img1 = Image.open(os.path.join(normal_path, normal_img))
img2 = Image.open(os.path.join(pneumonia_path, pneumonia_img))
axes[0].imshow(img1, cmap="gray")
axes[0].set_title("NORMAL")
axes[1].imshow(img2, cmap="gray")
axes[1].set_title("PNEUMONIA")

plt.show()
```

```python
print("Normal image size:", img1.size)
print("Pneumonia image size:", img2.size)
```

```python
import numpy as np
IMG_SIZE = 150 #It will resize all the images to 150x150 pixels

def load_images(folder, label):
    data = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        try:
            img = Image.open(img_path).convert("L") #convert to grayscale
            img = img.resize((IMG_SIZE, IMG_SIZE))
            img_array = np.array(img) / 255.0 #normalize pixel values to 0-1
            data.append((img_array, label))
        except Exception as e:
            print(f"Skipped {filename}: {e}")
    return data
```

```python
train_normal_data = load_images(os.path.join(base_dir, "train", "NORMAL"), label=0)
train_pneumonia_data = load_images(os.path.join(base_dir, "train", "PNEUMONIA"), label=1)

print(f"Loaded {len(train_normal_data)} normal images")
print(f"Loaded {len(train_pneumonia_data)} pneumonia images")
```

```python
val_normal_data = load_images(os.path.join(base_dir, "val", "NORMAL"), label=0)
val_pneumonia_data = load_images(os.path.join(base_dir, "val", "PNEUMONIA"), label=1)

test_normal_data = load_images(os.path.join(base_dir, "test", "NORMAL"), label=0)
test_pneumonia_data = load_images(os.path.join(base_dir, "test", "PNEUMONIA"), label=1)

print(f"VAL: {len(val_normal_data)} normal, {len(val_pneumonia_data)} pneumonia")
print(f"TEST: {len(test_normal_data)} normal, {len(test_pneumonia_data)} pneumonia")
```

```python
train_data = train_normal_data + train_pneumonia_data
val_data = val_normal_data + val_pneumonia_data
test_data = test_normal_data + test_pneumonia_data

import random
random.shuffle(train_data)
random.shuffle(val_data)
random.shuffle(test_data)

x_train = np.array([item[0] for item in train_data])
y_train = np.array([item[1] for item in train_data])

x_val = np.array([item[0] for item in val_data])
y_val = np.array([item[1] for item in val_data])

x_test = np.array([item[0] for item in test_data])
y_test = np.array([item[1] for item in test_data])

print("Shapes:")
print("x_train:", x_train.shape, "y_train:", y_train.shape)
print("x_val:", x_val.shape, "y_val:", y_val.shape)
print("x_test:", x_test.shape, "y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

#Flatten each 150x150 image into a single row of 22500 numbers
x_train_flat = x_train.reshape(x_train.shape[0], -1)
x_val_flat = x_val.reshape(x_val.shape[0], -1)

print("Flattened Shape:", x_train_flat.shape)

#Train logistic regression
baseline_model = LogisticRegression(max_iter=1000)
baseline_model.fit(x_train_flat, y_train)

#Evaluate on validation set
val_preds = baseline_model.predict(x_val_flat)
accuracy = accuracy_score(y_val, val_preds)

print(f"Baseline Validation Accuracy: {accuracy:.4f}")
print(classification_report(y_val, val_preds, target_names=["NORMAL", "PNEUMONIA"]))
```
